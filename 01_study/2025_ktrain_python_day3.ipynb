## 실습 문제 3

N명의 지원자(candidates) 중에서 K명을 뽑아 팀을 구성하려고 합니다. 가능한 모든 팀 조합을 리스트 형태로 반환하는 get_teams 함수를 작성하세요. 각 조합은 알파벳 순서로 정렬된 튜플 형태여야 합니다.

```python
from itertools import combinations, permutations

candidates = ["Alice", "Bob", "David", "Charlie"]

c = combinations(sorted(candidates), 2)
next(c),next(c),next(c),next(c)
```

```python

from itertools import combinations, permutations


def get_teams(candidates: list, k: int) -> list:
    return combinations(candidates, k)

input_data = [
    (["Alice", "Bob", "Charlie", "David"], 2),
    (["A", "B", "C"], 3),
    (["X", "Y"], 1),
    (["Z"], 2)
]

output_data = [
    [('Alice', 'Bob'), ('Alice', 'Charlie'), ('Alice', 'David'), ('Bob', 'Charlie'), ('Bob', 'David'), ('Charlie', 'David')],
    [('A', 'B', 'C')],
    [('X',), ('Y',)],
    []
]

for (candidates, k), expected in zip(input_data, output_data):
    result = get_teams(candidates, k)
    assert sorted(result) == sorted(expected), f"Failed: candidates={candidates}, k={k}, expected={expected}, got={result}"

print("Problem 1: All tests passed!")
```

## 실습 문제 4

달리기 경주에 참가하는 선수들(runners)이 있습니다. 이 선수들을 출발선에 세울 수 있는 모든 가능한 순서(순열)를 리스트 형태로 반환하는 get_race_lineups 함수를 작성하세요.

```python
from itertools import permutations

def get_race_lineups(runners: list) -> list:
    return list(permutations(runners))

input_data = [
    (["Leo", "Kai", "Sue"],),
    (["A", "B"],),
    (["Solo"],),
    ([],)
]

output_data = [
    [('Leo', 'Kai', 'Sue'), ('Leo', 'Sue', 'Kai'), ('Kai', 'Leo', 'Sue'), ('Kai', 'Sue', 'Leo'), ('Sue', 'Leo', 'Kai'), ('Sue', 'Kai', 'Leo')],
    [('A', 'B'), ('B', 'A')],
    [('Solo',)],
    [()]
]

for (runners,), expected in zip(input_data, output_data):
    result = get_race_lineups(runners)
    assert sorted(result) == sorted(expected), f"Failed: runners={runners}, expected={expected}, got={result}"

print("Problem 2: All tests passed!")
print(f"Failed: runners={runners}, expected={expected}, got={result}")
```

## 실습 문제 5

레스토랑에서 제공하는 여러 카테고리의 메뉴(categories)가 주어집니다. 각 카테고리에서 메뉴를 하나씩 선택하여 만들 수 있는 모든 코스 요리 조합을 반환하는 create_meal_courses 함수를 작성하세요

```python

import itertools


def create_meal_courses(*categories) -> list:
    return list(itertools.product(*categories))

input_data = [
    (["샐러드", "스프"], ["스테이크", "파스타"], ["아이스크림", "케이크"]),
    (["김밥"], ["라면", "우동"]),
    (["아메리카노"], ["치즈케이크"], []),
    (["짜장면"])
]

input_data = [
    (["샐러드", "스프"], ["스테이크", "파스타"], ["아이스크림", "케이크"]),
    (["김밥"], ["라면", "우동"]),
    (["아메리카노"], ["치즈케이크"], []),
    (["짜장면"],)
]

output_data = [
    [('샐러드', '스테이크', '아이스크림'), ('샐러드', '스테이크', '케이크'), ('샐러드', '파스타', '아이스크림'), ('샐러드', '파스타', '케이크'), ('스프', '스테이크', '아이스크림'), ('스프', '스테이크', '케이크'), ('스프', '파스타', '아이스크림'), ('스프', '파스타', '케이크')],
    [('김밥', '라면'), ('김밥', '우동')],
    [],
    [('짜장면',)]
]

for categories, expected in zip(input_data, output_data):
    result = create_meal_courses(*categories)
    assert sorted(result) == sorted(expected), f"Failed: categories={categories}, expected={expected}, got={result}"

print("Problem 3: All tests passed!")
```

## 실습 문제 6

여러 명의 직원(employees)이 순환 근무를 합니다. 총 N일(days) 동안의 일일 근무자를 배정하여 리스트로 반환하는 assign_daily_shifts 함수를 작성하세요. 근무는 직원 리스트 순서대로 순환됩니다.

```python
from itertools import cycle, islice

def assign_daily_shifts(employees: list, days: int) -> list:
    return list(islice(cycle(employees), days))

input_data = [
    (["철수", "영희", "민준"], 7),
    (["A", "B"], 4),
    (["Solo"], 5),
    (["X", "Y"], 0)
]

output_data = [
    ['철수', '영희', '민준', '철수', '영희', '민준', '철수'],
    ['A', 'B', 'A', 'B'],
    ['Solo', 'Solo', 'Solo', 'Solo', 'Solo'],
    []
]

for (employees, days), expected in zip(input_data, output_data):
    result = assign_daily_shifts(employees, days)
    assert result == expected, f"Failed: employees={employees}, days={days}, expected={expected}, got={result}"

print("Problem 4: All tests passed!")
```

## 실습 문제 7

문자열 데이터가 주어졌을 때, 연속으로 반복되는 문자를 (문자, 반복 횟수) 형태의 튜플로 묶어 리스트로 반환하는 compress_string 함수를 작성하세요.

```python

from itertools import groupby


def compress_string(data: str) -> list:
    return [(char, len(list(group))) for char, group in groupby(data)]

input_data = [
    ("AAABBCDDDD",),
    ("ABCA",),
    ("WWWWWW",),
    ("",)
]

output_data = [
    [('A', 3), ('B', 2), ('C', 1), ('D', 4)],
    [('A', 1), ('B', 1), ('C', 1), ('A', 1)],
    [('W', 6)],
    []
]

for (data,), expected in zip(input_data, output_data):
    result = compress_string(data)
    assert result == expected, f"Failed: data='{data}', expected={expected}, got={result}"

print("Problem 5: All tests passed!")
```
